## II. Matplotlib Case Study

```python
plt.plot([1,2,3,4,5]) # 실제 plotting을 하는 함수 # y = x + 1
plt.show() # plt를 확인하는 명령
```

```python
figsize = (10, 10)

plt.figure(figsize = figsize) # plotting을 할 도면을 선언

plt.plot([0, 1, 2, 3, 4]) # 실제 plotting을 하는 함수 # y = x
plt.show() # plt를 확인하는 명령
```

**2차함수 그래프 with plot()**

```python
# 리스트를 이용해서 1차함수 y=x를 그려보면:

plt.plot([0, 1, 2, 3, 4])
```
